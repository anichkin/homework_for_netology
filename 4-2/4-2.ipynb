```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
DATA_DIR = "C:/4-1/names" 
```

```python
names_by_year = {}
for year in range(1900, 2000):
    names_by_year[year] = pd.read_csv('{}/yob{}.txt'.format(DATA_DIR, year),
        names=['Name','Gender','Count']
    )
names_all = pd.concat(names_by_year, names=['Year', 'Pos'])
```

```python
name_dynamics = names_all.groupby([names_all.index.get_level_values(0), 'Name']).sum()
```

```python
name_dynamics.query('Name == ["Ruth", "Robert"]')
```

```python
name_dynamics.query('Name == ["Ruth", "Robert"]').unstack('Name')
```

```python
name_dynamics.query('Name == ["Ruth", "Robert"]').unstack('Name').plot()
```

```python
names_by_year = {}
for year in range(1900, 2005, 5):
    names_by_year[year] = pd.read_csv('{}/yob{}.txt'.format(DATA_DIR, year),
        names=['Name','Gender','Count']
    )
names_all = pd.concat(names_by_year, names=['Year', 'Pos'])
```

```python
name_dynamics = names_all.groupby([names_all.index.get_level_values(0), 'Name']).sum()
name_dynamics.query('Name == ["Ruth", "Robert"]').unstack('Name').plot.bar()
```

```python
names_by_year = {}
names_by_year[1950] = pd.read_csv('{}/yob{}.txt'.format(DATA_DIR, 1950),
        names=['Name','Gender','Count']
    )
names_all = pd.concat(names_by_year, names=['Year', 'Pos'])
names_all
```

```python
name_by_r = names_all[names_all.Name.str.match('R')]
name_by_r = name_by_r.sort_values(by='Count',ascending=False)
name_by_r
```

```python
names_for_pie = name_by_r.groupby('Name').sum().sort_values(by='Count', ascending=False).head(10)
names_for_pie
```

```python
names_for_pie.plot.pie(y='Count')
```
